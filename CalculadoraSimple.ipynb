```python
num1 = (int(input("ingresar número")))
num2 = (int(input("ingresar número")))
action = (str(input("Elija una opción: Add(a), Sub(b), Mult(m), Div(d) ->")))

print("The Result is", end=" " )
if action == "a":
   print(num1+num2)
elif action == "b":
   print(num1-num2)
elif action == "m":
   print(num1*num2)
else:
   print(num1/num2)
```
